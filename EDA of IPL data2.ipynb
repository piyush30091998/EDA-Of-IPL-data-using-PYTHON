```python
strike_rate_countries=data.groupby('COUNTRY')['SR-B'].mean()
```

```python
Top_5_countries= strike_rate_countries.sort_values(ascending=False).head(5)
```

```python
Top_5_countries = Top_5_countries.reset_index()
```

```python
sns.barplot(x='SR-B', y='COUNTRY', data=Top_5_countries, color='g')
plt.title('Top 5 Countries with Highest Average Strike Rate')
plt.show()
```

```python
# All-Rounder analysis
```

```python
all_rounders= data[(data['PLAYING ROLE'] == 'Allrounder')]
```

```python
sns.scatterplot(x='SR-B', y='SR-BL', data=all_rounders)

plt.xlabel('Batting Strike Rate')
plt.ylabel('Bowling Strike Rate')
plt.title('Batting vs Bowling Strike Rate for T20 All-rounders')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
sns.scatterplot(x='SR-B', y='SR-BL', data=all_rounders, hue='PLAYER NAME',sizes=(100,500),ax=ax)
# add the player names to the plot
for i, row in all_rounders.iterrows():
    ax.text(row['SR-B'], row['SR-BL'], row['PLAYER NAME'], ha='center', va='center', fontsize=10)
plt.xlabel('Batting strike rate')
plt.ylabel('Bowling strike rate')
plt.title('Batting vs Bowling strike rate of all rounders in IPL')
plt.show()
```

```python
# Sorting the data by percent difference in descending order and select the top 10 rows
```
